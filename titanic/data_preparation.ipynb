### Filtering specific rows from a DataFrame

```python
above_35 = titanic[titanic["Age"] > 35] # To select rows based on a conditional expression, use a condition inside the selection brackets [].
```

```python
above_35.head()
```

```python
titanic["Age"] > 35
```

```python
above_35.shape
```

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])] # equivalent to class_23 = titanic[(titanic['Pclass'] == 2) | (titanic["Pclass"] == 3)]
```

```python
class_23.head()
```

```python
age_no_na = titanic[titanic['Age'].notna()]; # The notna() conditional function returns a True for each row the values are not a Null value. 
```

```python
age_no_na.head()
```

```python
age_no_na.shape
```

### Selecting specific rows and columns from a DataFrame

```python
adult_names = titanic.loc[titanic['Age'] > 35, "Name"] # Selecting specific rows and/or columns using loc when using the row and column names.
```

```python
adult_names.head()
```

```python
titanic.iloc[9:25, 2:5] # rows 10 till 25 and columns 3 to 5
```

```python
titanic.iloc[0:3, 3] = "anonymous" # Selecting specific rows and/or columns using iloc when using the positions in the table
```

```python
titanic.head() # assigning new values to a selection based on loc/iloc
```
